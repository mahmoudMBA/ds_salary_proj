### parse salary

```python
gd['Salary Estimate'].value_counts()
```

If we have an value in our dependant veriable we drop that line

```python
gd=gd[gd['Salary Estimate']!='-1']
```

Remove (Glassdoor est.) from Salary Estimate

```python
salary=gd['Salary Estimate'].apply(lambda x : x.split("(")[0] )
```

Replace the 'K' and '$' by  ""

```python
salary=salary.apply(lambda x: x.replace("K","").replace('$',''))
```

Get the maximum,minimum and average salary and addd them to the data frame as integers

```python
gd['min_salary']=salary.apply(lambda x:int(x.split('-')[0]))
gd['max_salary']=salary.apply(lambda x:int(x.split('-')[1]))
gd['avg_salary']=(gd['min_salary']+gd['max_salary'])/2
```

```python
#gd[gd['Salary Estimate'].str.contains('Per Hour')]
```

### Clean company name

```python
gd[gd['Rating']==-1]
```

```python
gd['Company Name']=gd.apply(lambda x: x['Company Name'] if x['Rating']<0 else x['Company Name'][:-3],axis=1)
```

### Separate city and state

```python
gd['City']=gd['Location'].apply(lambda x:x.split(",")[0])
```

```python
#gd.State.value_counts()
```

Take the second part of location to create the state but if there is only one index in the list take the first index

```python
gd['State']=gd['Location'].apply(lambda x : x.split(',')[1] if len(x.split(','))>1 else x.split(',')[0]  )
```

Check for the values that we need to transform manually

```python
gd[gd['State'].str.len()>2]['State'].value_counts()
```

Replace the states that are not correct

```python
gd['State']=gd['State'].apply(lambda x: x.replace('Winnipeg','MB').replace('New York State','NY').replace('Toronto','ON').replace('Oregon','OR').replace('North Vancouver','BC').replace('New Jersey','NJ').replace('Nebraska','NE').replace('Edmonton','AB').replace(' Los Angeles','CA').replace('Richmond','VA').replace('California','CA').replace('Maryland','VA').replace('Virginia','VA').replace('Poland','Warsaw').replace('United States','Not Defined').replace('Remote','Not Defined'))
```

Replace " " by "" 

```python
gd['State']=gd['State'].apply(lambda x: x.replace(" ",""))
```

Create a list of US states and Canada states

```python
US_states = ["AL", "AK", "AZ", "AR", "CA", "CO", "CT", "DC", "DE", "FL", "GA", 
          "HI", "ID", "IL", "IN", "IA", "KS", "KY", "LA", "ME", "MD", 
          "MA", "MI", "MN", "MS", "MO", "MT", "NE", "NV", "NH", "NJ", 
          "NM", "NY", "NC", "ND", "OH", "OK", "OR", "PA", "RI", "SC", 
          "SD", "TN", "TX", "UT", "VT", "VA", "WA", "WV", "WI", "WY"]
Canada_states=["NL","PE","NS","NB","QC","ON","MB","SK","AB","BC","YT","NT","NU"]
```

Create a country field

```python
gd['Country']=gd.apply(lambda x: 'USA' if x['State'] in US_states or x['Location']=='United States' else('Canada' if x['State'] in Canada_states  else( 'Poland' if x['State']=='Warsaw' else 'Not Defined')),axis=1)
```

Check if the job position is at the Headquarter or not

```python
gd['At_hq']=gd.apply(lambda x: 1 if x['Headquarters']==x['Location'] else 0,axis=1)
```

### Parse Revenue

```python
gd['Revenue'].value_counts()
```

```python
revenue=gd['Revenue'].apply(lambda x: x.replace('Unknown / Non-Applicable','0').replace('-1','0').replace('$100 to $500 million (USD)','100-500').replace('$10+ billion (USD)','10000').replace('$50 to $100 million (USD)','50-100').replace('$10 to $25 million (USD)','10-25').replace('$1 to $2 billion (USD)','1000-2000').replace('$25 to $50 million (USD)','25-20').replace('$500 million to $1 billion (USD)','500-1000').replace('$2 to $5 billion (USD)','2000-5000').replace('$1 to $5 million (USD)','1-5').replace('$5 to $10 billion (USD)','5000-10000').replace('$5 to $10 million (USD)','5-10').replace('Less than $1 million (USD)','1'))
```

```python
gd['Min_revenue']=revenue.apply(lambda x: int(x.split('-')[0]))
gd['Max_revenue']=revenue.apply(lambda x:int(x.split('-')[1] if len(x.split('-'))>1 else x.split('-')[0]))
gd['Average_revenue']=(gd['Min_revenue']+gd['Max_revenue'])/2
```
